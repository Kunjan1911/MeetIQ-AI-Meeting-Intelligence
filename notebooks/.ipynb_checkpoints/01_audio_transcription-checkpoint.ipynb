# MeetIQ - Audio Transcription using Whisper

**Objective:** Convert a meeting audio file into text using OpenAI Whisper.

```python
import whisper
import os
```

```python
#load whisper model
model = whisper.load_model("base")
print("✅ Whisper Model Loaded Successfully!")
```

```python
output_folder = "../outputs"

os.makedirs(output_folder, exist_ok=True)

print("Output folder is ready.")
```

## Specify Audio File Path

```python
audio_path = "../sample_audio/meeting.mp4"

print(audio_path)
```

```python
import os

print(audio_path)
print(os.path.exists(audio_path))
```

## Transcribe the Meeting Audio

```python
result = model.transcribe(audio_path)
```

```python
print(result.keys())
```

```python
## Extract the Transcript
transcript = result["text"]

print(transcript)
```

```python
#Detect the Language
print("Detected Language:", result["language"])
```

```python
# Save the Transcript
output_file = os.path.join(output_folder, "meeting_transcript.txt")

with open(output_file, "w", encoding="utf-8") as file:
    file.write(transcript)

print("Transcript saved successfully!")
```

```python
# Read the Saved Transcript
with open(output_file, "r", encoding="utf-8") as file:
    saved_transcript = file.read()

print(saved_transcript)
```

```python
## Save Complete Transcription Result
import json

json_path = os.path.join(output_folder, "transcription_result.json")

with open(json_path, "w", encoding="utf-8") as file:
    json.dump(result, file, indent=4)

print("Complete transcription saved successfully!")
```
