# MeetSense AI - Sentiment Analysis & Meeting Analytics

## Objective

Analyze the overall sentiment of the meeting transcript and generate useful meeting statistics.

```python
import warnings

# Suppress and ignore all warnings globally
warnings.filterwarnings("ignore")
from transformers import pipeline
import pandas as pd
import re
```

```python
#Load Meeting Transcript
transcript_path = "../outputs/meeting_transcript.txt"

with open(transcript_path, "r", encoding="utf-8") as file:
    transcript = file.read()

print(transcript)
```

```python
# Load Sentiment Analysis Model
sentiment_pipeline = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)

print("Sentiment Model Loaded Successfully!")
```

```python
## Analyze Meeting Sentiment
sentiment = sentiment_pipeline(transcript[:512])[0]

print(sentiment)
```

```python
## Display Sentiment Result
sentiment_label = sentiment["label"]
confidence = round(sentiment["score"] * 100, 2)

print(f"Overall Sentiment : {sentiment_label}")
print(f"Confidence Score : {confidence}%")
```

```python
## Count Meeting Statistics

import re

# Split transcript into sentences
sentences = re.split(r'(?<=[.!?])\s+', transcript)

# Action keywords
action_keywords = [
    "will",
    "should",
    "need to",
    "complete",
    "finish",
    "prepare",
    "start",
    "integrate",
    "develop",
    "finalize",
    "perform",
    "update",
    "push",
    "submit",
    "review",
    "test"
]

# Count action items
action_items = []

for sentence in sentences:
    if any(keyword in sentence.lower() for keyword in action_keywords):
        action_items.append(sentence)

action_count = len(action_items)

word_count = len(transcript.split())
sentence_count = len(sentences)

print("Word Count:", word_count)
print("Sentence Count:", sentence_count)
print("Action Items:", action_count)
```

```python
## Calculate Meeting Productivity Score

score = 50

# Action Items
score += min(action_count * 8, 24)

# Sentiment
if sentiment_label == "POSITIVE":
    score += 15
elif sentiment_label == "NEGATIVE":
    score -= 15

# Meeting Length
if word_count > 100:
    score += 10

# Limit score between 0 and 100
score = max(0, min(score, 100))

print(f"Meeting Productivity Score: {score}/100")
```
