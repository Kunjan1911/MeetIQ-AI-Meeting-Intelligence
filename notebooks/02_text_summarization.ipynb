# MeetIQ - AI Meeting Summarization

**Objective:** Generate a concise and meaningful summary from the meeting transcript using a Transformer model.

```python
from transformers import pipeline
import os
import warnings

# Suppress all warnings
warnings.filterwarnings("ignore")
```

```python
# Load the Transcript
transcript_path = "../outputs/meeting_transcript.txt"

with open(transcript_path, "r", encoding="utf-8") as file:
    transcript = file.read()

print(transcript)
```

```python
summarizer = pipeline(
    "summarization",
    model="facebook/bart-large-cnn"
)

print(" Summarization Model Loaded Successfully!")
```

```python
#Generate Meeting Summary
summary = summarizer(
    transcript,
    max_length=120,
    min_length=40,
    do_sample=False
)

summary_text = summary[0]["summary_text"]

print(summary_text)
```

```python
summary_path = "../outputs/meeting_summary.txt"

with open(summary_path, "w", encoding="utf-8") as file:
    file.write(summary_text)

print("Summary Saved Successfully!")
```

```python
## Display Saved Summary
with open(summary_path, "r", encoding="utf-8") as file:
    print(file.read())
```
